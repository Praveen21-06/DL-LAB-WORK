PART 1 — Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

PART 2 — Dataset (AND / OR)

```python
X = [(0,0),(0,1),(1,0),(1,1)]
Y_AND = [0,0,0,1]
Y_OR  = [0,1,1,1]
```

PART 3 — Sigmoid function

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

PART 4 — Training function

```python
def train_gate(Y, lr):

    w1, w2, b = 0.2, -0.3, 0.4
    losses = []

    for epoch in range(50):
        total_loss = 0

        for (x1,x2), y in zip(X,Y):

            z = w1*x1 + w2*x2 + b
            y_pred = sigmoid(z)

            loss = 0.5*(y_pred-y)**2
            total_loss += loss

            dL_dy_pred = y_pred - y
            dy_pred_dz = y_pred*(1-y_pred)
            dL_dz = dL_dy_pred * dy_pred_dz

            w1 -= lr*(dL_dz*x1)
            w2 -= lr*(dL_dz*x2)
            b  -= lr*dL_dz

        losses.append(total_loss)
        print(f"Epoch {epoch+1} Loss = {total_loss}")

    return losses
```

PART 5 — Learning rates

```python
learning_rates = [0.1,0.5,1]
```

PART 6 — AND gate loss plot

```python
plt.figure()
for lr in learning_rates:
    print("\nAND Gate LR =", lr)
    losses = train_gate(Y_AND, lr)
    plt.plot(losses,label=f"lr={lr}")

plt.title("AND Gate Loss")
plt.legend()
plt.show()
```

PART 7 — OR gate loss plot

```python
plt.figure()
for lr in learning_rates:
    print("\nOR Gate LR =", lr)
    losses = train_gate(Y_OR, lr)
    plt.plot(losses,label=f"lr={lr}")

plt.title("OR Gate Loss")
plt.legend()
plt.show()
```
