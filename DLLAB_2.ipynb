Step 1 — Change the epochs and observe the output (AND gate)

```python
import numpy as np

x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])

epochs = 5
w, b = np.zeros(2), 0

for epoch in range(epochs):
    for x, y in zip(x_list, y_list):
        z = np.dot(x, w) + b
        y_pred = 1 if z >= 0 else 0
        E = y - y_pred
        print("error:", E, "<- at epoch", epoch)
        w += E * x
        b += E
        print(w, b)
```

Step 2 — Use random weights and bias, observe convergence

```python
import numpy as np

x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])

epochs = 5
w, b = np.random.randn(2), np.random.randn()

for epoch in range(epochs):
    for x, y in zip(x_list, y_list):
        z = np.dot(x, w) + b
        y_pred = 1 if z >= 0 else 0
        E = y - y_pred
        print("error:", E, "<- at epoch", epoch)
        w += E * x
        b += E
        print(w, b)
```

Step 3 — Add logic to stop when error becomes zero for all inputs

```python
import numpy as np

x_list = np.array([[0,0],[0,1],[1,0],[1,1]])
y_list = np.array([0,0,0,1])

epochs = 10
w, b = np.random.randn(2), np.random.randn()

for epoch in range(epochs):
    total_error = 0

    for x, y in zip(x_list, y_list):
        z = np.dot(x, w) + b
        y_pred = 1 if z >= 0 else 0
        E = y - y_pred
        total_error += abs(E)

        w += E * x
        b += E
        print("error:", E, "<- at epoch", epoch, w, b)

    if total_error == 0:
        print("Training converged at epoch", epoch)
        break
```
