STEP 1 — Import Library

```python
import numpy as np
```

STEP 2 — Define Sigmoid Function

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

STEP 3 — Define Perceptron Function

```python
def perceptron(x1, x2, w1, w2, b):
    z = w1*x1 + w2*x2 + b
    return sigmoid(z)
```

STEP 4 — Define Input Combinations

```python
inputs = [
    (0,0),
    (0,1),
    (1,0),
    (1,1)
]
```

STEP 6 — Compute Table

```python
print("x1  x2   OR   NAND   XOR")
print("--------------------------------")
for x1, x2 in inputs:

    # OR Gate
    *
    or_out = perceptron(x1, x2, 1, 1, -0.5)
    or_out = int(or_out >= 0.5)

    # NAND Gate
    nand_out = perceptron(x1, x2, -1, -1, 1.5)
    nand_out = int(nand_out >= 0.5)

    # AND Gate
    and_out = perceptron(x1, x2, 1, 1, -1.5)
    and_out = int(and_out >= 0.5)

    # XOR = OR AND NAND
    xor_out = perceptron(or_out, nand_out, 1, 1, -1.5)
    xor_out = int(xor_out >= 0.5)

    print(f"{x1}   {x2}    {or_out}     {nand_out}      {xor_out}")
```
